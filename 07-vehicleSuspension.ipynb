# **Suspensão automotiva**
### Modelagem de uma suspensão automotiva

A modelagem a seguir utiliza Equações Diferenciais para representar o evento físico da suspensão.

Design sem nome (10).png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
m = 3
b = 1
k = 2
Fext = 10
```

$$ ma = F_{ext} - kx - bv$$
$$ a = \frac{F_{ext}}{m} - \frac{k}{m}x - \frac{b}{m}v$$
$$ \frac{d^2 x}{dt^2} = \frac{F_{ext}}{m} - \frac{k}{m}x - \frac{b}{m}\frac{dx}{dt}$$




A equação acima é de segunda ordem então, reduz-se elas a duas equações diferenciais ordinárias de primeira ordem, abaixo:

## $\frac{d x}{dt} = v $

## $ \frac{d v}{dt} = \frac{F_{ext}}{m} - \frac{k}{m}x - \frac{b}{m}v$



Onde:
* F: força  (N)
* v: velocidade (m/s)
* x: deslocamento (m)
* a: aceleração (m/s²)



```python
def dXdt(t, X, m, b, k, Amp):
    x, v = X
    Ftempo  = Amp#*np.sin(t)
    temp = [v, Ftempo/m - k/m*x -b/m*v]
    return temp
```

A função tempo é similar à distribuição delta de Dirac, os valores dos parâmetros utilizados são:


*   m = 3
*   b = 1
*   k = 2
*   Fext = 10



```python
tt = np.linspace(0, 50, 10000)
fig, ax = plt.subplots()
fig.set_size_inches(15, 4)
for kk in range(-4, 15, 5):
    x_0 = kk
    v_0 = 0
    X_0 = (x_0, v_0)
    sol = odeint(dXdt, y0=X_0, t=tt, tfirst=True, args=(m, b, k, Fext))
    ax.plot(tt, sol[:,0], label='$x=$'+str(x_0))
ax.legend()
ax.set_xlabel("$t$")
ax.set_ylabel("$x(t)$")
ax.grid()
fig.show()
```

A resposta ao impulso do sistema é estável. As massas se movem de forma senoidal. Essas ondas senoidais decaem e retornando ao equilíbrio.
