```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,0.05,100)
y = 2**(1/2) * 200 * np.sin(2 * np.pi * 60 * x)
print("A média é =",np.mean(y))
```

```python
fig,ax = plt.subplots()
ax.plot (x , y , 'b-',linewidth=1)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
ax.set_title('Gráfico 01')
plt.show()
```

```python
t = np.linspace(0,0.05,300)
w = (2**(1/2) * 220 * np.sin(2 * np.pi * 60 * t))**2
media = np.mean(w)
print("A média é =",media)
square = media ** (1/2)
print("A raiz quadrada é =",square)
```

```python
fig,ax = plt.subplots() 
ax.plot (t , w , 'r-',linewidth=1)
plt.xlabel('Eixo y')
plt.ylabel('Eixo x')
ax.set_title('Gráfico 02')
plt.show()
```
