```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#theta=pi/3
V=38.5 #velocidade em m/s
g= 9.8 #gravidade m/s²
t=np.linspace(0,20,1000)

angle = np.pi/12

while angle < np.pi/2:
    x=V*np.cos(angle)*t
    y=V*np.sin(angle)*t+(0.5*-9.8*t**2)
    plt.ylim([0,80])
    plt.xlim([0,170])
    plt.plot(x,y)
    angle+=np.pi/12

plt.xlabel("alcance")
plt.ylabel("altura")
plt.title('Gráfico de lançamento oblíquo')
plt.show()
```
