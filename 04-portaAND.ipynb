O seguinte problema trata sobre neurônios.Um neurônio recebe entradas, sinais elétricos, processa e executa um sinal elétrico de saída. Nossos neurônios não possuem um comporttamento de meio-termo, logo se a combinação de entradas exceder um limite estipulado, um sinal de saída é executado. Se a combinação fior menor que certo limite, então o neurônio não produz nenhuma saída.

```python
import numpy as np
```

**As entradas são binárias (0 ou 1) e dadas por vetores. A mesma coisa é para as saídas, exatamente o mesmo número de pesos.**

CodeCogsEqn.svg



*   *'s' seria a pré-ativação*
*   *P seria os pesos*






```python
entradas = np.array([[0,0],[0,1], [1,0], [1,1]])
saidas = np.array([0,0,0,1])
pesos = np.array([0.0, 0.0])
taxaAprendizagem = 0.3
```

Depois de ter a soma, aplicou-se uma função de ativação. A função de ativação, pode ser aproximada para uma função degrau, pois seu comportamento gráfico é como de uma escada.

CodeCogsEqn (1).svg

Então, se a entrada for maior ou igual a 0, a saída produzida é 1. Caso contrário, a saída produzida será 0. Este é o modelo matemático para um único neurônio, a unidade mais fundamental para uma rede neural .

Este código é sobre a porta AND, que quando insere-se (1,1) a saída é 1. Os neurônios são como as portas apresentadas.Recebem 2 entradas binárias e produzem uma única saída binária.
Abaixo, no primeiro bloco de código têm-se a função ativação que retorna 1 se a entrada for maior ou igual a 0 e caso contrário, retorna 0.
A seguir, uma função executa uma entrada retornando uma saída e calcula-se o produto interno aplicado a uma função de ativação.
E por último, foi criada uma função, dadas as entradas e saídas desejadas. Continuou-se atualizando os pesos várias vezes e iterando por todo o conjunto treinar.  Após isso, foi calculado o erro e produzido a regra de atualização.

```python
def stepFunction(soma):
    if (soma >= 1):
        return 1
    return 0

def calculaSaida(registro):
    s = np.dot(registro, pesos)
    return stepFunction(s)

def treinar():
    erroTotal = 1
    while (erroTotal != 0):
        erroTotal = 0
        for i in range(len(saidas)):
            saidaCalculada = calculaSaida(np.asarray(entradas[i]))
            erro = saidas[i] - saidaCalculada
            erroTotal =  erroTotal + erro
            for j in range(len(pesos)):
                pesos[j] = pesos[j] + (taxaAprendizagem * entradas[i][j] * erro)
                print('Peso atualizado: ' + str(pesos[j]))
        print('Total de erros: ' + str(erroTotal))
```

```python
treinar()
print('Rede neural treinada')
print(calculaSaida(entradas[0]))
print(calculaSaida(entradas[1]))
print(calculaSaida(entradas[2]))
print(calculaSaida(entradas[3]))

print(pesos)
```
